# **Type of Analysis**

**Descriptive Analysis:**
Summarizes the dataset using statistical measures like mean, median, and trends. It helps in understanding overall port performance, cargo volumes, and yearly patterns. Visualizations like bar charts and line graphs are used to present the data clearly.

**Diagnostic Analysis:**
Focuses on identifying the reasons behind patterns and trends in the data. It compares variables such as region, growth rate, and cargo volume to understand performance differences. This helps explain why certain ports perform better than others.

**Predictive Analysis:**
Uses historical data to forecast future trends and performance. Techniques like trend analysis or simple regression help predict cargo volume and growth rates. It provides an idea of future outcomes based on past data.

**Prescriptive Analysis:**
Provides recommendations based on insights gained from the analysis. It suggests actions to improve port performance, efficiency, and growth. This helps in making informed business and operational decisions.

**Stage 3 – EDA and Visualizations**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("https://raw.githubusercontent.com/swapnamarri2711-msw/PORT/refs/heads/main/global_port_dataset_6000_rows.csv")


num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

print("Numerical:", num_cols)
print("Categorical:", cat_cols)
```

**Univariate Analysis → distribution of single variables (countplot, histogram, boxplot)**

**COUNT PLOT**

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df['Country'])
plt.xticks(rotation=45)
plt.title("Count of Ports by Country")
plt.show()
```

Insights:

Shows frequency distribution

Identify:

Most common country/port

Imbalance in data

Example:

If China/India dominates → dataset is skewed

If few countries dominate → not globally balanced

**HISTOGRAM (Numerical Variables)**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Cargo_Tons'], bins=30, kde=True)
plt.title("Distribution of Cargo Volume")
plt.show()
```

Insights:

Shows distribution shape

Normal → symmetric
Right skew → many small values, few large ports

Example:

If right-skewed → few ports handle massive cargo

Helps detect outliers & skewness

**BOXPLOT (Outliers Detection)**

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Cargo_Tons'])
plt.title("Boxplot of Cargo Volume")
plt.show()
```

Insights:
Detect:

Median
Spread (IQR)
Outliers
Example:

Many dots above box → high-performing ports

Large spread → high variability in operations

**BAR PLOT (Categorical vs Numerical)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/swapnamarri2711-msw/PORT/refs/heads/main/global_port_dataset_6000_rows.csv")


plt.figure(figsize=(10,5))
sns.barplot(x=df['Country'], y=df['Cargo_Tons'])
plt.xticks(rotation=45)
plt.title("Average Cargo Volume by Country")
plt.show()
```

Insights:

Compares numeric values across categories

Shows:

Which country performs best
Which has lowest activity

Example Insight:

Some countries dominate → global trade concentration

Large variation → economic imbalance

**CORRELATION HEATMAP**

```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Insights:

Values range:
 -1 to +1

+1 → strong positive

-1 → strong negative

0 → no relation

Interpretation:

The heatmap represents the pairwise correlation between numerical variables.

Color intensity indicates strength:

Dark red → strong positive correlation

Dark blue → strong negative correlation

Light color → weak/no correlation

**Analys about the sales by region**

**Bar Plot – Total Sales by Region**

```python
plt.figure(figsize=(8,5))
region_sales = df.groupby('Region')['Growth_Rate_%'].sum().reset_index()

sns.barplot(x='Region', y='Growth_Rate_%', data=region_sales)
plt.title("Total Sales by Region")
plt.show()
```

Interpretation:

This plot shows the total sales aggregated for each region.

The height of each bar represents overall contribution of that region to total sales.

 Insights:

The region with the highest bar is the top-performing region.

Regions with lower sales indicate potential areas for improvement.

A large gap between regions suggests uneven sales distribution.

**Pie Chart**

```python
region_sales.set_index('Region')['Growth_Rate_%'].plot(
    kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Sales Distribution by Region")
plt.ylabel("")
plt.show()
```

Interpretation:

Displays the percentage contribution of each region to total sales.

Insights:

Helps identify dominant regions in overall sales.

If one region occupies a large portion → market concentration.

Balanced slices → even distribution across regions.

**Line Plot**

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Year', y='Growth_Rate_%', hue='Region', data=df)
plt.title("Sales Trend by Region")
plt.show()
```

Interpretation:

Shows how sales change over time across regions.

Insights:

Upward trend → growth in that region.

Decline → possible business issues.

Crossing lines → changing regional performance over time.
